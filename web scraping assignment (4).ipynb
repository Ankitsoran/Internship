Q1Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:

A) Rank B) Name C) Artist D) Upload date E) Views

```python
#Activating the chrome browser

driver=webdriver.Chrome(r"C:/Users/hp/Downloads/chromedriver.exe") 
time.sleep(4)
#Getting youtube wikipedia page
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
time.sleep(6)
```

```python
# Creating empty list for scaping the data

Rank = []
Name = []
Artist = []
Upload_Date = []
Views = []
```

```python
# Scraping Rank of the videos
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]"):
        Rank.append(i.text)
except NoSuchElementException:
    Rank.append("_")
    
# Scraping Name of the videos
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]"):
        Name.append(i.text)
except NoSuchElementException:
    Name.append("_")
    
# Scraping Artist of the video
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]"):
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append("_")

# Scraping Uploaded date of the video
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]"):
        Upload_Date.append(i.text)
except NoSuchElementException:
    Upload_Date.append("_")
    
# Scraping Views of videos
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]"):
        Views.append(i.text)
except NoSuchElementException:
    Views.append("_")
    
# Creating DataFrame from scraped data
Wiki = pd.DataFrame({})
Wiki['Rank']=Rank
Wiki['Name']=Name
Wiki['Artist']=Artist
Wiki['Upload Date']=Upload_Date
Wiki['Views']=Views

# Removing stray numbers from Name column
Wiki.Name = Wiki.Name.apply(lambda x:x[:-4].strip('"'))
Wiki
```

```python
# webpage opening
driver.get("https://github.com/")

time.sleep(5)
```

```python
#clicking explore tab
expl_tag = driver.find_element_by_xpath("//div[1]/header/div/div[2]/nav/ul/li[4]/details/summary")
expl_tag.click()
    
time.sleep(2)
#trnd
trend_tag = driver.find_element_by_xpath("//ul[@class='list-style-none f5 pb-1']/li[2]/a")
trend_tag = driver.find_element_by_xpath("//div[@class='d-flex flex-wrap flex-items-center flex-justify-center flex-md-justify-start text-center text-md-left']/a[3]")
trend_tag.click()
    

```

```python
# vreating variable to store the data
title = []
desc = []
contri = []
lang = []
url = []

#scraping title
for h in driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a"):
    title.append(h.text[h.text.find("/")+1:])

# sraping language used
loop = range(1,26)
for i in loop:
    try:
        a = driver.find_element_by_xpath('//*[@id="js-pjax-container"]/div[3]/div/div[2]/article['+str(i)+']/div[2]/span[1]/span[2]')
        lang.append(a.text)
    except NoSuchElementException:
        lang.append('-')

# scraping contribution count
for j in driver.find_elements_by_xpath("//div[@class='f6 color-fg-muted mt-2']/a[2]"):
    contri.append(j.text)
    
# scraping title and url
for k in driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a"):
    url.append(k.get_attribute('href'))
# scraping title and language
for l in driver.find_elements_by_xpath("//div[@class='f6 color-fg-muted mt-2']/span/span[2]"):
    lang.append(l.text) 
# scraping title and  Repository description    
for m in driver.find_elements_by_xpath("//p[@class='col-9 color-fg-muted my-1 pr-4']"):
    desc.append(m.text)
```

```python
print(len(url),len(contri),len(title),len(lang),len(desc))
lang=lang[0:22]
title=title[0:22]
contri=contri[0:22]
url=url[0:22]
```

```python
github = pd.DataFrame({ 'URL':url,'Repository title': title,
                      'Contributord count': contri,
                      'Language Used': lang,'Description':desc})
github
```
